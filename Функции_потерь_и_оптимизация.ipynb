```python
from sklearn import datasets
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
#from jupyterthemes import jtplot
#jtplot.style()
```

```python
iris = datasets.load_iris()
print(iris.target_names)
```

```python
#Выбор только двух цветков
X = []
Y = []
for i, unit in enumerate(iris.target):
    if unit != 0:
        X.append(iris.data[i])
        Y.append(unit-1)

X = np.asarray(X)
Y = np.asarray(Y)
print(Y)
```

```python
def predict(theta, X):
    predict = theta[0] + theta[1] * X[:, 0] + theta[2] * X[:, 1] + theta[3] * X[:, 2] + theta[4] * X[:, 3]
    sigm = 1. / (1 + np.exp(-predict))
    return sigm
```

```python
#Градиентный спуск
EPOCHS = 100
LEARNING_RATE = 0.05
N = 100
#lr = 0.05
#theta = []
theta = np.random.normal(scale=5, size=N)
#theta = [-0.66556365, -1.46787087, -0.21254648, -0.14877584,  2.22796437]
print(theta)
grad_loss = []
for _ in range(EPOCHS):
    sigm = predict(theta, X)
    theta[0] -= LEARNING_RATE * np.sum(sigm - Y)/len(sigm)
    theta[1] -= LEARNING_RATE * np.sum((sigm - Y) * X[:, 0])/len(sigm)
    theta[2] -= LEARNING_RATE * np.sum((sigm - Y) * X[:, 1])/len(sigm)
    theta[3] -= LEARNING_RATE * np.sum((sigm - Y) * X[:, 2])/len(sigm)
    theta[4] -= LEARNING_RATE * np.sum((sigm - Y) * X[:, 3])/len(sigm)
    loss = - np.mean(np.log(sigm) * Y + np.log(1 - sigm) * (1 - Y))
    print(theta)
    grad_loss.append(loss)
```

```python
error = 0
for i, item in enumerate(np.around(predict(theta, X))):
    if item != Y[i]:
        error += 1
error
```

```python
# Надстройка nesterov momentum
EPOCHS = 100
LEARNING_RATE = 0.01
N = 100
#theta = np.random.normal(scale=5, size=N)
theta = [-0.66556365, -1.46787087, -0.21254648, -0.14877584,  2.22796437]
vel_pred = np.zeros(5)
vel = np.zeros(5)
print(theta)
nest_loss = []
gamma = 0.975
for _ in range(EPOCHS):
    sigm = predict(theta, X)
    loss = - np.mean(np.log(sigm) * Y + np.log(1 - sigm) * (1 - Y))
    nest_loss.append(loss)
    sigm = predict(theta - gamma * vel_pred, X)
    vel[0] = (gamma * vel_pred[0] + LEARNING_RATE * np.sum(sigm - Y))/len(sigm)
    vel[1] = (gamma * vel_pred[1] + LEARNING_RATE * np.sum((sigm - Y) * X[:, 0]))/len(sigm)
    vel[2] = (gamma * vel_pred[2] + LEARNING_RATE * np.sum((sigm - Y) * X[:, 1]))/len(sigm)
    vel[3] = (gamma * vel_pred[3] + LEARNING_RATE * np.sum((sigm - Y) * X[:, 2]))/len(sigm)
    vel[4] = (gamma * vel_pred[4] + LEARNING_RATE * np.sum((sigm - Y) * X[:, 3]))/len(sigm)
    theta -= vel
    print(theta)
    vel_pred = vel
```

```python
error = 0
for i, item in enumerate(np.around(predict(theta, X))):
    if item != Y[i]:
        error += 1
error
```

```python
# Надстройка RMSProp
EPOCHS = 100
LEARNING_RATE = 0.01
N = 100
#theta = np.random.normal(scale=5, size=N)
theta = [-0.66556365, -1.46787087, -0.21254648, -0.14877584,  2.22796437]
e_sq_grad = np.zeros(5)
grad = np.zeros(5)
print(theta)
rmsp_loss = []
gamma = 0.975
eps = 0.00000001
for _ in range(EPOCHS):
    sigm = predict(theta, X)
    loss = - np.mean(np.log(sigm) * Y + np.log(1 - sigm) * (1 - Y))
    rmsp_loss.append(loss)
    grad[0] = np.sum(sigm - Y)/len(sigm)
    grad[1] = np.sum((sigm - Y) * X[:, 0])/len(sigm)
    grad[2] = np.sum((sigm - Y) * X[:, 1])/len(sigm)
    grad[3] = np.sum((sigm - Y) * X[:, 2])/len(sigm)
    grad[4] = np.sum((sigm - Y) * X[:, 3])/len(sigm)
    
    e_sq_grad = gamma * e_sq_grad + (1 - gamma)  * grad ** 2
    
    theta -= LEARNING_RATE * grad / np.sqrt(e_sq_grad + eps)
    print(theta)
```

```python
error = 0
for i, item in enumerate(np.around(predict(theta, X))):
    if item != Y[i]:
        error += 1
error
```
